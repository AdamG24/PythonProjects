```python
autos.loc[autos["odometer_km"].between(0, 30000), "odometer_km_group"] = "1"
autos.loc[autos["odometer_km"].between(30001, 60000), "odometer_km_group"] = "2"
autos.loc[autos["odometer_km"].between(60001, 90000), "odometer_km_group"] = "3"
autos.loc[autos["odometer_km"].between(90001, 120000), "odometer_km_group"] = "4"
autos.loc[autos["odometer_km"].between(120001, 150000), "odometer_km_group"] = "5"
autos.loc[autos["odometer_km"] > 150001, "odometer_km_group"] = "6"
```

```python
group_price_mean = {}

for i in autos["odometer_km_group"].unique():
    group = autos[autos["odometer_km_group"] == i]
    mean_price = group["price"].mean()
    group_price_mean[i] = int(mean_price)
    
group_price_mean
```

Lower the odometer, higher the price

```python
mapping_dict_damage = {"nein":"no",
                       "ja":"yes",
                      }
autos["unrepaired_damage"] = autos["unrepaired_damage"].map(mapping_dict_damage)
autos["unrepaired_damage"].value_counts()
```

```python
damaged = {}

car = autos["brand"].unique()
for i in car:
    selected = autos[(autos["brand"] == i)& (autos["unrepaired_damage"]=="yes")]
    damaged[i] = selected["price"].mean()
damaged
```

```python
non_damaged = {}

car = autos["brand"].unique()
for i in car:
    selected = autos[(autos["brand"] == i)& (autos["unrepaired_damage"]=="no")]
    non_damaged[i] = selected["price"].mean()
non_damaged
```

```python
damaged_series = pd.Series(damaged)
non_damaged_series = pd.Series(non_damaged)

damaged_vs_non_damaged = pd.DataFrame(damaged_series, columns=['mean_price_damaged'])
damaged_vs_non_damaged["mean_price_non_damaged"] = non_damaged_series

damaged_vs_non_damaged["difference"] = damaged_vs_non_damaged["mean_price_non_damaged"] - damaged_vs_non_damaged["mean_price_damaged"]
damaged_vs_non_damaged
```
