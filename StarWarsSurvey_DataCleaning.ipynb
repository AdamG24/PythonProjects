```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

```python
yes_no = {'Yes': True, 'No': False}
for col in ['Have you seen any of the 6 films in the Star Wars franchise?', 'Do you consider yourself to be a fan of the Star Wars film franchise?']:
    star_wars[col] = star_wars[col].map(yes_no)
    

star_wars.head()
```

```python
import numpy as np

movie_mapping = {"Star Wars: Episode I  The Phantom Menace": True, np.nan: False, "Star Wars: Episode II  Attack of the Clones": True, "Star Wars: Episode III  Revenge of the Sith": True, "Star Wars: Episode IV  A New Hope": True, "Star Wars: Episode V The Empire Strikes Back": True, "Star Wars: Episode VI Return of the Jedi": True}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
    
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1", 'Unnamed: 4': 'seen_2', 'Unnamed: 5': 'seen_3', 'Unnamed: 6': 'seen_4', 'Unnamed: 7': 'seen_5', 'Unnamed: 8': 'seen_6'})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1", 'Unnamed: 10': 'ranking_2', 'Unnamed: 11': 'ranking_3', 'Unnamed: 12': 'ranking_4', 'Unnamed: 13': 'ranking_5', 'Unnamed: 14': 'ranking_6'})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

Our theory and assumptionm proved to be supported by the data we just analyzed. Based off the rankings the sample provided, the 5th movie, Empire Strikes Back, received higher average score than any other movie in the series. To be exact, the rank of the movies went as the following: 5, 6, 4, 1, 2, 3.
